**名称：** CTR

**来源：** 20240109-东吴证券-换手率切割刀 CTR 因子 Cutlets of Turnover Rate——换手率的异质信念  


**作者：**rc

**构造方法：**  
1、每月月初开盘后，回看股票过去 20 个交易日内的隔夜换手率和隔夜收益率，计
算每个交易日的隔夜聪明（收益/换手），取隔夜聪明钱最小的四天

2、再取这四天前一个交易日的日内换手率，取均值后做市值中性化，得到新因子

```python
p=ff.run(CTR.loc[:,'20200210':'20230410'],period=1,positions=100,fees=0)
```

**名称：** JumpCTR

**来源：** 20240109-东吴证券-换手率切割刀 CTR 因子 Cutlets of Turnover Rate——换手率的异质信念  


**作者：**rc

**构造方法：**  
1、每月月底, 取股票过去 20 个交易日的日内换手率, 按照次日隔夜聪明钱从低到高排序；

2、 依次记为 $t r_1^{\prime}, t r_2^{\prime} \ldots \ldots t r_{20}^{\prime}$, 等分为 5 组

3、取隔夜隔夜聪明钱最小的三天的换手率, 加上本月最后一个交易日的日内换手率,取这四天的换手率平均值后市值中性化得到新因子, 抢跑 CTR (JumpCTR)。

```python
p=ff.run(JumpCTR.loc[:,'20200210':'20230410'],period=1,positions=100,fees=0)
```
